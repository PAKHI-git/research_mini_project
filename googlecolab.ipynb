```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    "Year": [2021, 2022, 2023, 2024],
    "Publications": [5, 9, 14, 20]
})

plt.plot(data["Year"], data["Publications"], marker="o")
plt.title("Research Publications Growth")
plt.xlabel("Year")
plt.ylabel("No. of Publications")
plt.show()
```
